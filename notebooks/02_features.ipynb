```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/bitcoin_processed.csv", parse_dates=['date'])

print("Datos cargados. Columnas disponibles:")
print(df.columns)

# ANLISIS 1: Matriz de Correlacin
features = ['price', 'sma_7', 'sma_30', 'rsi', 'price_lag_1', 'price_lag_7', 'volatility']

plt.figure(figsize=(10, 8))
correlation_matrix = df[features].corr()

# Dibujar mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Mapa de Calor de Correlaciones (Qu predice mejor el precio?)")
plt.show()

# ANLISIS 2: El poder del Lag 
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['price_lag_1'], y=df['price'], alpha=0.5)
plt.title("Correlacin: Precio de Ayer vs Precio de Hoy")
plt.xlabel("Precio Ayer (Lag 1)")
plt.ylabel("Precio Hoy")

plt.grid(True)
plt.show()

```
